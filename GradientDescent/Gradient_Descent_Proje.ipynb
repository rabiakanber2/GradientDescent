## Verileri Olu≈ütur

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100,1)
X
```

```python
error = np.random.randn(100,1)
error
```

```python
y = 4 + 3* X + error    #errola grafiƒüin doƒüru ≈üeklinde √ßƒ±ktƒ± vermesini engelledik datayƒ± daƒüƒ±ttk.
y   
```

```python
# veriyi √ßizelim

plt.plot(X,y,'b.')
plt.xlabel("$x$", fontsize = 18)
plt.ylabel("$y$", rotation=0, fontsize = 18)
_ =plt.axis([0,2,0,15])
```

## Lineer Regresyonun Analitik √á√∂z√ºm√º

$$ y = WX $$

Burada bulmaya √ßalƒ±≈ütƒ±ƒüƒ±mƒ±z $W$ vekt√∂r√ºd√ºr.

Matrisler ile bu √ß√∂z√ºm:

$$W = (X^TX)^{-1}X^Ty$$

```python
# X'e bias column ekle  1lerden olu≈üan kolon ekler
X_b = np.c_[np.ones((100,1)),X]
```

```python
# katsayƒ± tahminlerini hesapla -> numpy ile √ß√∂z
# Yukarƒ±daki formildeki W yi hesaplƒ±yoruz.

w_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(w_best)
```

**Yeni data geldiƒüinde tahmin yap:**

```python
# yeni data -> X_new

X_new = np.array([[0],[2]])
X_new
```

```python
# bias column ekle

X_new_b = np.c_[np.ones((2,1)),X_new]
X_new_b
```

```python
# tahmin yap
# elimizde optimum w deƒüerleri var -> w_best
#Python predict() function -> , eƒüitilmi≈ü model temelinde veri deƒüerlerinin etiketlerini tahmin etmemizi saƒülar.

y_predict = X_new_b.dot(w_best)
y_predict
```

Bu lineer √ß√∂z√ºm i√ßin grafik √ßiz:

```python
# tahmin doƒürusu
plt.plot(X_new, y_predict, 'r-')

#b√ºt√ºn data
plt.plot(X, y, 'b.')

plt.xlabel("$x_1$", fontsize = 18)
plt.ylabel("$y$", rotation=0, fontsize = 18)
plt.axis([0,2,0,15])
plt.show()
```

### Gradient Descent

≈ûimdi aynƒ± i≈ülemi Gradient Descent ile yapalƒ±m:

```python
# Maliyet (Cost) Hesaplama Fonksiyonu

def cal_cost(W, X, y):
    """
    X ve y i√ßin Cost(Maliyet) hesaplar.
    Parametreler:
    W = Katsayƒ± vekt√∂r√º -> (p, 1)
    X = Girdi vekt√∂r√º   -> (n, p)
    y = Ger√ßek y'ler    -> (n, 1)
    Burada:
    n: Toplam veri adedi (satƒ±r sayƒ±sƒ±)
    p: X i√ßindeki deƒüi≈üken adedi (S√ºtun sayƒ±sƒ±)
    
    D√∂n√º≈ü:
    cost = hesaplanan maliyet deƒüeri
    """
    
    n = len(y)
    
    # √∂nce tahmini hesapla -> ùë¶ÃÇ = X W
    # X -> (n,p)
    # W -> (p, 1)
    # ùë¶ÃÇ -> (n, 1)
    tahmin = X.dot(W)
    
#a ve b vekt√∂rlerinin nokta √ßarpƒ±mƒ±nƒ± d√∂nd√ºr√ºr.
    
    #cost function deƒüerini hesapla -> J
    cost = (1/2*n) * np.sum(np.square(tahmin-y))
    
#np.square -> Bu matematiksel fonksiyon, kullanƒ±cƒ±nƒ±n dizideki her bir elemanƒ±n kare deƒüerini hesaplamasƒ±na yardƒ±mcƒ± olur.
    #maliyet (cost) deƒüerini d√∂n
    return cost
    
```

```python
# Gradient Descent Fonksiyonu
# ùõº  : learning rate
# iteration=100 d√∂ng√ºy√º ka√ß defa yapacaƒüƒ±mƒ±zƒ± belirtiriz.
def gradient_descent(X, y, W, learning_rate=0.01, iterations=100):
    """
    Gradient Descent ugulama fonksiyonu
    parametreler:
    X = X matrisi (bias unit eklenmi≈ü hali, yani 1'lerden olu≈ümu≈ü ilk s√ºtun)
    y = y vekt√∂r√º
    W = katsayƒ± vekt√∂r√º (w'lardan olu≈ümu≈ü)
    learning_rate = learning rate: alpha(√∂ƒürenme katsayƒ±sƒ±)
    iterations = toplam d√∂ng√º sayƒ±sƒ±
    D√∂n√º≈ü:
    * W vekt√∂r√ºn√ºn son hali
    * Maliyet Listesi (cost history)
    * W vekt√∂r√ºn√ºn listesi (weight history)
    """
    
    n=len(y)   #toplam veri adeti sayƒ±sƒ±
    cost_history = np.zeros(iterations)        
    w_history = np.zeros((iterations, 2))
# maliyet ge√ßmi≈üini yazdƒ±rma
# w 2 boyutlu olduƒüu i√ßin 2 yazdƒ±k.
    for it in range(iterations):
        tahmin = np.dot(X, W)
        W = W - (1/n) * learning_rate * (X.T.dot((tahmin - y)))
        w_history[it,:] = W.T
        cost_history[it] = cal_cost(W, X, y)
    return W, cost_history, w_history
        
```

```python
# learning rate
lr = 0.01

#iterasyon sayƒ±sƒ±
n_iter = 1000

#baslangƒ±c i√ßin W vekt√∂r√º 
W = np.random.randn(2,1)

#X vekt√∂r√ºne bias column ekle
X_b = np.c_[np.ones((len(X),1)),X]

#gradient descent √ßalƒ±≈ütƒ±r
W, cost_history, w_history = gradient_descent(X_b, y, W, lr, n_iter)

print("W0: {:0.3f}".format(W[0][0]))
print("W1: {:0.3f}".format(W[1][0]))
print("Final cost/MSE: {:0.3f}".format(cost_history[-1]))
```

**Iterasyonlarƒ± √áizelim:**

```python
fig,ax = plt.subplots(figsize=(12,8))

ax.set_ylabel("J(W)")
ax.set_xlabel("Iterations")
_=ax.plot(range(n_iter),cost_history,"b.")
```

**Gradient Descent'in nasƒ±l √ßalƒ±≈ütƒ±ƒüƒ±nƒ± fonksiyon ile g√∂relim:**

```python
# Gradient Descent'i √ßizen fonksiyon

def plot_GD(n_iter, lr, ax, ax1=None):
    """
    n_iter = iterayon sayƒ±sƒ±
    lr = Learning Rate
    ax = Gradient Descent √ßizmek i√ßin axis
    ax1 = cost_history vs Iterations grafiƒüi i√ßin axis
    """
    
    _=ax.plot(X, y, 'b.')
    W = np.random.randn(2,1)
    
    tr = 0.1
    cost_history = np.zeros(n_iter)
    
    for i in range(n_iter):
        pred_prev = X_b.dot(W)
        
        W, h, _ = gradient_descent(X_b, y,W,lr, 1)
        pred = X_b.dot(W)
        
        cost_history[i] = h[0]
        if ((i % 25 == 0 )):
            _ = ax.plot(X, pred,'r-', alpha=tr)
            if tr < 0.8:
                tr = tr+0.2
    if not ax1==None:
        _ = ax1.plot(range(n_iter),cost_history,'b.')
```

```python
fig = plt.figure(figsize=(30,25))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

#iterasyon ve learning rate listesi
it_lr =[(2000,0.001),(500,0.01),(200,0.5),(100,0.1)]

count = 0
for n_iter, lr in it_lr:
    count+= 1
    
    ax =fig.add_subplot(4,2,count)
    count+=1
    
    ax1 =fig.add_subplot(4,2,count)
    
    ax.set_title("lr:{}".format(lr))
    ax1.set_title("Iterations:{}".format(n_iter))
    
    plot_GD(n_iter,lr,ax,ax1)
```

### Stochastic Gradient Descent

```python
# Stochastic Gradient Descent Fonksiyonu

def stochastic_gradient_descent(X, y, W, learning_rate=0.01, iterations=100):
    """
    Stochastic Gradient Descent uygulama fonksiyonu.
    Parametreler:
    X = X Matrisi (bias unit eklenmi≈ü hali,yani 1'lerden olu≈ümu≈ü ilk s√ºtun)
    y = y vekt√∂r√º 
    W = Katsayƒ± vekt√∂r√º (w'lardan olu≈ümu≈ü)
    learning_rate = learning rate: alpha (√∂ƒürenme katsayƒ±sƒ±)
    iterations = toplam d√∂ng√º sayƒ±sƒ±
    D√∂n√º≈ü:
    *W vekt√∂r√ºn√ºn son hali 
    *Maliyet Listesi (cost history)
    *W vekt√∂r√ºn√ºn listesi (weight history)
    """
    
    n = len(y)
    cost_history = np.zeros(iterations)
    
    for it in range(iterations):
        cost = 0
        #her seferinde rasgele bir X_i deƒüeri se√ßip maliyet hesaplama
        for i in range(n):
            #rasgele bir deƒüer al
            rand_ind = np.random.randint(0,n)
            
            X_i = X[rand_ind,:].reshape(1,X.shape[1])
            y_i = y[rand_ind].reshape(1,1)
            
            #tek X_i i√ßin tahmin 
            tahmin = np.dot(X_i,W)
            
            #tek X_i i√ßin katsayƒ± tahmini 
            W =  W - (1/n) * learning_rate * (X_i.T.dot((tahmin - y_i)))
            
            # tek X_i i√ßin hesaplanan cost'u cost deƒüi≈ükenine ekle
            cost += cal_cost(W, X_i,y_i)
            
        #bu iterasyon i√ßin cost deƒüerini cost_history'ye ekle
        cost_history[it] = cost
        
    return W, cost_history
    
```

```python
# learning rate
lr = 0.5

#iterasyon sayƒ±sƒ±
n_iter = 50

#baslangƒ±c i√ßin W vekt√∂r√º 
W = np.random.randn(2,1)

#X vekt√∂r√ºne bias column ekle
X_b = np.c_[np.ones((len(X),1)),X]

#gradient descent √ßalƒ±≈ütƒ±r
W, cost_history = stochastic_gradient_descent(X_b, y, W, lr, n_iter)

print("W0: {:0.3f}".format(W[0][0]))
print("W1: {:0.3f}".format(W[1][0]))
print("Final cost/MSE: {:0.3f}".format(cost_history[-1]))
```


≈ûimdi SGD'yi grafik √ºzerinde g√∂relim:

```python
fig,ax = plt.subplots(figsize=(12,8))

ax.set_ylabel("{J(W)}",rotation=0)
ax.set_xlabel("{Iterations}")
_=ax.plot(range(n_iter),cost_history,"b.")
```

G√∂rd√ºƒü√ºn√ºz gibi normal Gradient Descent gibi lineer azalmadƒ±.

Bazen arttƒ± bazen azaldƒ±. √á√ºnk√º rasgele se√ßtik noktalarƒ±.

Ama nihayetinde o da azalarak giden bir eƒüim izledi.

### Mini-Batch Gradient Descent

≈ûimdi SGD'yi bir de partiler (batch) √ºzerinden yapalƒ±m. Yani Mini-Batch Stochastic Gradient Descent.

```python
# Mini-Batch Stochastic Gradient Descent Fonksiyonu

def minibatch_stochastic_gradient_descent(X, y, W, learning_rate=0.01, iterations=100, batch_size=20):
    """
    Stochastic Gradient Descent uygulama fonksiyonu.
    Parametreler:
    X = X Matrisi (bias unit eklenmi≈ü hali, yani 1'lerden olu≈ümu≈ü ilk s√ºtun)
    y = y vekt√∂r√º
    W = Katsayƒ± vekt√∂r√º (w'lardan olu≈ümu≈ü)
    learning_rate = learning rate: alpha (√∂ƒürenme katsayƒ±sƒ±)
    iterations = toplam d√∂ng√º sayƒ±sƒ±
    D√∂n√º≈ü:
    * W vekt√∂r√ºn√ºn son hali
    * Maliyet Listesi (cost history)
    * W vekt√∂r√ºn√ºn listesi (weight history)
    """
    
    n = len(y)
    cost_history = np.zeros(iterations)
    # toplam ka√ß adet batch olacak
    n_batches = int(n / batch_size)
    
    for it in range(iterations):
        cost = 0
        # X ve y'leri karƒ±≈ütƒ±r (veri sƒ±ralarƒ± rasgele olsun)
        indices = np.random.permutation(n)
        X = X[indices]
        y = y[indices]
        
        # her seferinde rasgele bir X_i partisi alƒ±p maliyet hesaplayacaƒüƒ±z
        for i in range(0, n, batch_size):
            # rasgele X_i ve y_i partileri
            X_i = X[i:i+batch_size]
            y_i = y[i:i+batch_size]
        
            # bias column ekle (1'ler s√ºtunu)
            x_i = np.c_[np.ones(len(X_i)), X_i]
        
            # X_i partisi i√ßin tahmin
            tahmin = np.dot(X_i, W)

            # X_i partisi i√ßin katsayƒ± deƒüi≈üimi
            W = W - (1/n) * learning_rate * (X_i.T.dot((tahmin - y_i)))

            # X_i partisi i√ßin hesaplanan cost'u cost deƒüi≈ükenine ekle
            cost += cal_cost(W, X_i, y_i)
        
        # bu iterasyon i√ßin cost deƒüerini cost_history'ye ekle
        cost_history[it] = cost
    
    return W, cost_history
```

```python
# learning rate
lr = 0.1

# iterasyon sayƒ±sƒ±
n_iter = 200

# ba≈ülangƒ±√ß i√ßin W vekt√∂r√º
W = np.random.randn(2,1)

# X vekt√∂r√ºne bias column ekle
X_b = np.c_[np.ones((len(X),1)),X]

# gradient descent √ßalƒ±≈ütƒ±r
W, cost_history = minibatch_stochastic_gradient_descent(X_b, y, W, lr, n_iter)


print('W0: {:0.3f}'.format(W[0][0]))
print('W1: {:0.3f}'.format(W[1][0]))
print('Final cost/MSE: {:0.3f}'.format(cost_history[-1]))
```

≈ûimdi Mini-Batch SGD'yi grafik √ºzerinde g√∂relim:

```python
fig,ax = plt.subplots(figsize=(10,8))

ax.set_ylabel('{J(W)}',rotation=0)
ax.set_xlabel('{Iterations}')

_=ax.plot(range(n_iter), cost_history, 'b.')
```

G√∂rd√ºƒü√ºn√ºz gibi, SGD'yi teker teker deƒüil de Batch yani partiler √ºzerinden yaparsak daha istikrarlƒ± bir seyir izleyerek optimum Cost deƒüerine yakla≈ütƒ±.
